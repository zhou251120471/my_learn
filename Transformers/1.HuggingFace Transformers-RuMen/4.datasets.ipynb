```python
train_datasets=load_dataset('madao33/new-title-chinese',split='train')
train_datasets
```

```python
train_datasets=load_dataset('madao33/new-title-chinese',split='train[:100]')
train_datasets
```

```python
train_datasets=load_dataset('madao33/new-title-chinese',split=['train[:50%]','validation[:10%]'])
train_datasets
```
